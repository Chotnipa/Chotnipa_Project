```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time

def run4():
    df3 = df222
    # Explicitly create an empty DataFrame
    result_df_list = []

    # Use a single WebDriver instance
    driver = webdriver.Chrome()

    for link in df3['Link']:
        try:
            # Open the URL
            driver.get(link)
            time.sleep(0.05)
            
            # Get the page source
            page_source = driver.page_source
   

            # Parse the HTML with BeautifulSoup
            soup = BeautifulSoup(page_source, 'html.parser')

            # Find the element with class "odd"
            odd_element = soup.find('tr', class_='odd')
            odd_text_content = odd_element.get_text(strip=True) if odd_element else None

            # Find the element with class "text_title"
            text_title_element = soup.find('span', class_='text_title')
            text_title_content = text_title_element.get_text(strip=True) if text_title_element else None
            time.sleep(0.05)
            # Debug prints
            #print(f"Link: {link}")
            #print(f"Odd Element Text: {odd_text_content}")
            #print(f"Text Title Element Text: {text_title_content}")

            # Create a temporary DataFrame for the current link
            temp_df = pd.DataFrame({'รายละเอียด': [text_title_content], 'อัตราการเข้าพักเฉลี่ย': [odd_text_content]})

            # Append the temporary DataFrame to the list
            result_df_list.append(temp_df)

        except Exception as e:
            print(f"Error processing {link}: {str(e)}")

    # Close the WebDriver
    driver.quit()

    result_df = pd.concat(result_df_list, ignore_index=True)

    result_df['อัตราการเข้าพักเฉลี่ย'] = result_df['อัตราการเข้าพักเฉลี่ย'].str.replace('อัตราการเข้าพักเฉลี่ย (OR)', '', regex=False)
    # Print or use the new DataFrame
    extracted_info = pd.DataFrame()

    extracted_info = result_df['รายละเอียด'].str.extract(r'สถานการณ์ท่องเที่ยวภายในประเทศจังหวัด (.+?)ประจำเดือน (.+?) (\d{4})')

    # Renaming columns
    extracted_info.columns = ['จังหวัด', 'เดือน', 'ปี']

    # Combining the extracted information with the original DataFrame
    extracted_info['อัตราการเข้าพักเฉลี่ย'] = result_df['อัตราการเข้าพักเฉลี่ย']

    # Displaying the result
    return extracted_info


df444 = run4()
df444
```
